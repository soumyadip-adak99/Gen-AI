# Task 2: Categories (set)


```python
# 1: from products list, create a set of categories called categories_set
categories = ["Electronics", "Personal care", "Fashion", "Electronics", "Personal care"]
categories_set = set(categories)

# 2: adding a new category to the set and show that duplicates are ingnored
categories_set.add("Books")
categories_set.add("Electronics")
print("Adding categories: ", categories_set)

# 3: show how to check whether a category exists in the set
present_category = "Electronics" in categories_set
print("Category is present: ", present_category)

# (optional): show total number of unique categories in a set
print("Total Unique categories: ", len(categories_set))
```

# Task 3: Product Pricing (Dictionaries)


```python
# 1: create a dictionary price_dict where keys are product names and values are price. include at least 6 entries
price_dict = {
    "Laptop": 70000,
    "T-shirt": 300,
    "Mobile": 30000,
    "Shampoo": 200,
    "Shop": 40,
}

# 2: Add new product
# update the price of and existing product
price_dict["T-shirt"] = 500
print("Update price", price_dict)

# remove product by name if the product if exits
remove_item = "Book"

if remove_item in price_dict:
    price_dict.pop(remove_item)
    print(f"{remove_item} product remove successfully.")
else:
    print(f"{remove_item} not found.")

print(price_dict)

# 3: Print the average price of all products
total_price = sum(price_dict.values())
avarage_price = total_price / len(price_dict)
print("Avarage of all product price: ", avarage_price)

# (optional): Print the product with both maximum and minimum price
max_product = max(price_dict, key=price_dict.get)
min_product = min(price_dict, key=price_dict.get)
print(
    f"Product with maximum price:\nProduct Name: {max_product}\nProduct price: {price_dict[max_product]}"
)
print(
    f"Product with minimum price:\nProduct Name: {min_product}\nProduct price: {price_dict[min_product]}"
)
```

## Task 4: Combination Operations

```python
# 1. create a list of tuples named catalog: (product_name, price, category)
product_names = list(price_dict.keys())
categories_for_products = categories[:len(product_names)]

catalog = []
for i in range(len(product_names)):
    name = product_names[i]
    cat = categories_for_products[i]
    price = price_dict[name]
    catalog.append((name, price, cat))

# 2. create category_to_products dictionary
category_to_products = {}
for name, price, cat in catalog:
    if cat not in category_to_products:
        category_to_products[cat] = []
    category_to_products[cat].append(name)

# 3. print products from the largest category (without lambda)
if category_to_products:
    max_category = None
    max_count = 0

    for cat in category_to_products:
        count = len(category_to_products[cat])
        if count > max_count:
            max_count = count
            max_category = cat

    print(f"Category with maximum products: {max_category}")
    print("Products:", category_to_products[max_category])
else:
    print("No products available.")

```
