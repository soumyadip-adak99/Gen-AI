### Task 1: Discount Rules (if/elif/else)

```python
order_amount = float(input("Enter amount: "))

if order_amount < 0:
    print("Order amount cannot be negetive.")
    exit()

if order_amount >= 2000:
    discount_amount = 0.15
elif order_amount >= 1500:
    discount_amount = 0.10
elif order_amount >= 1000:
    discount_amount = 0.07
else:
    discount_amount = 0.0

discount = order_amount * discount_amount
subtotal = order_amount - discount

tax = subtotal * 0.5
final_amount = subtotal + tax

print("Origianl amount: ", order_amount)
print("Discount: ", discount_amount)
print("Sub total amount after discount: ", subtotal)
print("Tax (5%): ", tax)
print("Final total amount: ", final_amount)
```

## Task 2: Process Multiple orders (for loop)

```python
orders = [1200, 2500, 800, 1750, 3000]

total_revenue = 0
discounted_order = 0

print("Order amount -> Discount -> Final amount")

for order_amount in orders:
    if order_amount >= 2000:
        discount_amount = 0.15
    elif order_amount >= 1500:
        discount_amount = 0.10
    elif order_amount >= 1000:
        discount_amount = 0.7
    else:
        discount_amount = 0.0
        
    discount = order_amount * discount_amount
    final_amount = order_amount - discount_amount
    
    if discount > 0:
        discounted_order += 1
    
    total_revenue  += final_amount
    
    print(f"{order_amount} -> {discount} -> {final_amount}")

print(f"Total revenue: {total_revenue}")
print(f"Total discounted product: {discounted_order}")
    
```

### Task 3: User menu (while loop + break/continue)

```python
order_list = []
total_amount = 0

while True:
    print("\nAdd order amount press 1")
    print("Show all orders press 2")
    print("Quit press 3")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        order = float(input("Enter amount: "))
        order_list.append(order)
    elif choice == 2:
        total_amount = 0

        for order_amount in order_list:
            if order_amount >= 2000:
                discount_rate = 0.15
            elif order_amount >= 1500:
                discount_rate = 0.10
            elif order_amount >= 1000:
                discount_rate = 0.07
            else:
                discount_rate = 0.0

            discount = order_amount * discount_rate
            total_amount += order_amount - discount

        print("All orders:", order_list)
        print("Total price after discount:", total_amount)
    elif choice == 3:
        print("Exiting program...")
        break
    else:
        print("Please choose correct option.")
```

### Loop control with condition (break & continue)

```python
daliy = [200, 150, 0, 400, 50, -1, 300]
total_revenue = 0

for i in daliy:
    if i == -1:
        print("Currupted data, stop processing")
        break

    if i == 0:
        print("No seals today")
        continue
    total_revenue += i
    print("Running total sales: ", total_revenue)

print("Final Total revenue: ", total_revenue)
```
