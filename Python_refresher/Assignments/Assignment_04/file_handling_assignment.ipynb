### Task 1: Write Sales Records to a File

```python
# 1: create a list of sales amounts
sales = [1200, 450, 980, 1500, 3000]

# 2: open file in write mode
file = open("sales_data.txt", "w")

# 3: write each sale on a new line
for s in sales:
    file.write(str(s) + "\n")

# 4: close file
file.close()

# 5: reopen the file and print its contents
file = open("sales_data.txt", "r")
content = file.read()
print(content)
file.close()
```

### Task 2: Read File in Different Ways

```python
# 1: read the entire file using .read() and print it.
with open("sales_data.txt", "r") as file:
    content = file.read()
    print("Full file content: ")
    print(content)
```

```python
# 2: read the first line using .readline().
with open("sales_data.txt", "r") as file:
    first_line = file.readline()
    print(f"First line: {first_line}")
```

```python
# 3: read all lines using .readlines() and convert them into a list of integers.
with open("sales_data.txt", "r") as file:
    lines = file.readlines()

sales = [int(line.strip()) for line in lines]
print(sales)
```

### Task 3: Append new Sales

```python
# 1: append these new sales to the same file
new_val = [5000, 2500, 1700]

with open("sales_data.txt", "a") as file:
    for s in new_val:
        file.write(str(s) + "\n")
```

```python
# 2: reopen and print the update file
with open("sales_data.txt", 'r') as file:
    data = file.read()
    print(data)
```

```python
# 3: print total numbers of lines after appending
with open("sales_data.txt", 'r') as file:
    lines = file.readlines()
    print("Total lines: ", len(lines))
```

### Task 4: Generate Summary report from file

```python
# 1: read all sales values from sales_data.txt
with open("sales_data.txt", "r") as file:
    lines = file.readlines()

# 2: convert them into integers
sales = [int(line.strip()) for line in lines]

# 3: calculate and print
print(f"Total sales: {sum(sales)}")
print(f"Highest Sale: {max(sales)}")
print(f"Lowest Sale: {min(sales)}")
print(f"Average Sale: {(sum(sales)/len(sales))}")
```
