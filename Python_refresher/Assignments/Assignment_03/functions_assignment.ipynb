### Task 1: Basic Function: Price After Discount

```python
def apply_discount(price, discount_percent=None):
    # if discount_percent is missing
    if discount_percent is None:
        discount_percent = 5

    # ensure discount never exceeds 60%
    if discount_percent > 60:
        discount_percent = 60

    discount_amount = price * (discount_percent / 100)
    final_price = price - discount_amount
    return final_price


print(apply_discount(1000, 10))
print(apply_discount(500))
```

### Task 2: Recursive Function: Factorial Utility

```python
def factorial(n):
    if n < 0: 
        print("factorial not defined for negative numbers")
        return None
    
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
print(factorial(0))
print(factorial(-3))
```

### Task 3: Lambda Function: GST Calculator

```python
gst = lambda price: price + (0.18 * price)
print(gst(100))

# otional: Create another lambda to compute final price after GST + discount together
final_price = lambda price, discount: (price + 0.18 * price) - (price * discount / 100)
print(final_price(1000, 10))
```

### Task 4: using map(): Apply GST to List of Prices

```python
prices = [100, 250, 400, 1200, 50, 200, 850]

gst = lambda price: price + (0.18 * price)

price_after_gst = list(map(gst, prices))

print(f"Origianl Price list: {prices}")
print(f"Prices after adding GST: {price_after_gst}")
```

### Task 5: using filter(): Filter Expensive Products

```python
prices = [100, 250, 400, 1200, 50, 2000, 850]

# prices filter greater than 500
list_1 = list(filter(lambda price: price > 500, prices))

# price filter less than or equal to 500
list_2 = list(filter(lambda price: price <= 500, prices))

print(f"Prices greater than 500: {list_1}")
print(f"Prices less than or equal to 500: {list_2}")
```

### Task 6: Combined Utility Function

```python
def process_price(prices):
    # apply 10% discount using map + lambda
    discount_price = list(map(lambda price: price - (price * 0.10), prices))

    # filter prices above 300
    filtered_prices = list(filter(lambda price: price > 300, discount_price))

    return discount_price, filtered_prices


result = process_price([100, 500, 900, 50, 750])

print(f"Discount prices: {result[0]}")
print(f"Filtered prices (>300): {result[1]}")
```

### Task 7: Mini Problem: Menu using functions

```python
def add_price(price_list, price):
    price_list.append(price)


def get_average_price(price_list):
    return sum(price_list) / len(price_list)


def get_max_price(prices_list):
    return max(prices_list)


price_list = []

while True:
    choice = input(
        "Menu:\n1 -> Add price\n2 -> Show average price\n3 -> Show highest price\nq -> Quit\nChoose: "
    ).strip()

    if choice == "1":
        p = float(input("Enter price: ").strip())
        add_price(price_list, p)
        print(f"Added: {p}")
    elif choice == "2":
        if price_list:
            print(f"Average price: {get_average_price(price_list)}")
        else:
            print("No prices available.")
    elif choice == "3":
        if price_list:
            print(f"Highest price: {get_max_price(price_list)}")
        else:
            print("No prices available.")
    elif choice.lower() == "q":
        print("Quitting.")
        break
    else:
        print("Invalid option.")
```
