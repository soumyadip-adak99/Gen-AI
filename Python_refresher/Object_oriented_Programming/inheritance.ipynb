### In python have 3 types of Inheritance: 
1. Single Inheritance
2. Multilevel Inheritance
3. Multiple Inheritance

### Single Inheritance

```python
class Avaneger:
    def __init__(self, name, power):
        self.name = name
        self.power = power

    def introduction(self):
        print(f"Hello my name is: {self.name} my power is {self.power}")


class Ironman(Avaneger):
    def __init__(self, name, power, weapon):
        super().__init__(name, power)
        self.weapon = weapon

    def weapon_details(self):
        print(f"I use {self.weapon}")


if __name__ == "__main__":
    iron = Ironman("Irongman", "Genius Intellect", "Electron")
    iron.introduction()
    iron.weapon_details()
```
