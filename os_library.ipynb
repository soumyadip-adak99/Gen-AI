os library is able to intreact with our operating system and can let us work with multiple Directories and Files

```python
## getcwd () -> it is capable of providing us with the idea about the current working directory
```

```python
import os 

current_path = os.getcwd()
print(current_path)
```

```python
## mkdir() -> stand for make directory
import os 

new_path =  "New_Folder"
os.mkdir(new_path)
print(f"The new directory {new_path} is create successfully")
```

```python
# listdir() -> Listing all of the directories present
import os 
all_directory = os.listdir(".")
print(all_directory)
```

```python
# joining 2 paths togerther
import os

dir_name = "Folder_data"
file_name = "File_data"
full_path = os.path.join(dir_name, file_name)
print(full_path)
```

```python
import os

dir_name = "Folder_data"
file_name = "File_data.txt"
current_complete_path = os.path.join(os.getcwd(), dir_name, file_name)
print(current_complete_path)
```

```python
# os.exists() -> check if the path is present is present or not
import os

give_path = "***"

if os.path.exists(give_path):
    print(1)
else:
    print(0)
```

```python
# os.path.isFile() --> this cheks that whether if the current path is a file or not
# os.path.isdir() --> this chekcs that whether if the current path is the directory is not

import os

path = 'Total_data.txt'
if os.path.isfile(path):
    print("Present")
elif os.path.isdir(path):
    print("The current path belongs to a directory")
else:
    print("Absent")
```

```python
# the comlete path <---> absolute path
# os.path.abspath()

import os

file_name = "print.py"
path = os.path.abspath(file_name)
print(path)
```
